```python
# while循环学习

# ======================
# 一、while循环基本语法
# ======================
"""
语法结构：
while 条件表达式:
    循环体（缩进4空格）

执行流程：
1. 先判断条件表达式是否为True
2. 如果为True，执行循环体
3. 循环体执行完后，回到步骤1重新判断条件
4. 直到条件表达式为False时，退出循环
"""

```

```python
# ======================
# 二、基础示例：计数循环
# ======================
print("===== 示例1：打印1-5的数字 ====")
num = 1  # 初始化计数器
while num <= 5:
    print(f"当前数字：{num}")
    num += 1  # 计数器自增（关键！避免死循环）

```

```python
# ======================
# 三、条件控制循环
# ======================
print("\n===== 示例2：累加直到超过100 ====")
sum_total = 0
current = 1
while sum_total < 100:
    sum_total += current
    print(f"累加{current}，当前总和：{sum_total}")
    current += 1
```

```python
# ======================
# 四、用户输入验证
# ======================
print("\n===== 示例3：输入偶数验证 ====")
while True:
    input_str = input("请输入一个偶数：")
    if input_str.isdigit():
        user_input = int(input_str)
        if user_input % 2 == 0:
            print(f"成功输入：{user_input}")
        else:
            print("输入错误，需要偶数")
    else:
        print("输入无效，请输入数字")
```

```python
# for循环学习

# ======================
# 一、for循环基本语法
# ======================
"""
语法结构：
for 变量 in 可迭代对象:
    循环体（缩进4空格）

执行流程：
依次从可迭代对象中取出元素赋值给变量，执行循环体，直到所有元素处理完毕
常见可迭代对象有：
字符串、列表、元组、字典、集合、range对象等
"""


# ======================
# 二、基础示例：遍历可迭代对象
# ======================
print("===== 示例1：遍历字符串 ====")
text = "Python"
for char in text:
    print(f"当前字符：{char}")

print("\n===== 示例2：遍历列表 ====")
numbers = [10, 20, 30, 40]
for num in numbers:
    print(f"列表元素：{num}")
```

```python
# ======================
# 三、range命令的使用
# ======================
"""
range(stop): 生成0到stop-1的整数序列
range(start, stop): 生成start到stop-1的整数序列
range(start, stop, step): 按步长step生成整数序列
"""
print("\n===== 示例3：使用range生成序列 ====")

# 生成0-4的序列（5个元素）
for i in range(5):
    print(f"range(5)的第{i+1}个数：{i}")

# 生成2-5的序列（2,3,4）
print("\nrange(2,6)的数：", end="")
for i in range(2, 6):
    print(i, end=" ")

# 生成1-9的奇数序列（1,3,5,7,9）
print("\n\nrange(1,10,2)的奇数：", end="")
for i in range(1, 10, 2):
    print(i, end=" ")
```

```python
# ======================
# 四、enumerate命令的使用
# ======================
"""
enumerate(可迭代对象)：返回(索引, 元素)的元组序列
常用于需要同时获取索引和元素的场景
"""
print("\n\n===== 示例4：使用enumerate遍历 ====")
fruits = ["苹果", "香蕉", "橙子"]
for index, fruit in enumerate(fruits):
    print(f"第{index+1}个水果：{fruit}")

```

```python


# ======================
# 五、综合练习：计算平均分
# ======================
print("\n===== 示例5：计算考试平均分 ====")
scores = [85, 92, 78, 90, 88]
sum_score = 0
for score in scores:
    sum_score += score
average = sum_score / len(scores)
print(f"班级平均分：{average:.2f}")
```
